```python
import pandas as pd
import numpy as np
import datetime
import timeit
```

```python
arr = pd.read_csv('household_power_consumption.txt', sep=';', low_memory=False).values
```

```python
#preparing data
arr = arr[arr[:,2]!='?']
dt=arr[:,0] + np.array([' ']*len(arr)) + arr[:,1]
dt = pd.to_datetime(dt)   
arr[:,0] = dt.date
arr[:,1] = dt.time
#arr = np.delete(arr, 1, 1)

for i in range(2,arr.shape[1]):
    arr[:,i] = arr[:,i].astype(float)
    
```

```python
#task 4
idx=np.sort(np.random.choice(arr.shape[0], 500000,replace=False))
arr4 = arr[idx,:]

print(np.mean(arr4[:,6]), np.mean(arr4[:,6]), np.mean(arr4[:,8]))
```

```python
#function for task 5
def p5(a):
    frame = a[(a[:,1]>datetime.time(18))&
              (a[:,2]>6)&
              (a[:,7]>=a[:,6])&
              (a[:,7]>=a[:,8])]
    booleans = [True if ((i+1)%3==0 and (i+1)<len(frame)/2) or 
                ((i+1)%4==0 and (i+1)>=len(frame)/2) else False for i in range(len(frame))]
    return frame[booleans]
```

```python
#task 5
p5(arr4)
```
